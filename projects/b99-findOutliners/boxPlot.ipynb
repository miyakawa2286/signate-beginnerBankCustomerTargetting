```python
import os
print(os.getcwd())
```

```python
import pandas as pd 
dpath_to_data = '../../data/'

train = pd.read_csv(os.path.join(dpath_to_data,'train.csv'))
test = pd.read_csv(os.path.join(dpath_to_data,'test.csv'))
```

```python
train_mean = train.mean()
train_std = train.std()
train_ = (train-train_mean)/train_std
test_ = (test-train_mean)/train_std
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set()
```

```python
fig,axs = plt.subplots(nrows=2,figsize=(10,10))
train_.boxplot(ax=axs[0])
test_.boxplot(ax=axs[1])
```
